1. สมมติให้ $I$ เป็นรูปภาพ การปรับ contrast บน $I$ จะใช้สูตร
$$ I' = \text{Truncate}(F(I-127.5)+127.5) $$
โดยที่ 
* $I'$ เป็นรูปใหม่ที่ได้   
* $F$ มีค่าระหว่าง $0$ กับ $1$  
* $\text{Truncate}(x)=0$ ถ้า $x<0$
* $\text{Truncate}(x)=255$ ถ้า $x>255$  
จงเขียนฟังก์ชัน `contrast(img, F)` สำหรับการปรับ contrast ของรูป `img` ด้วยสูตรข้างบน โดยที่ $F$ มีค่าเท่ากับ `F`

```
def contrast(img, F):

    def truncate(i):
        if i < 0: return 0
        elif i > 255: return 255
        else: return i

    out = img.copy()
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            out[i,j] = truncate(F * (img[i,j]-127.5) + 127.5)
    
    return out
```

```
plt.imshow(contrast(img, 0.7), cmap='gray', vmin=0, vmax=255);
```

```
plt.imshow(contrast(img, 0.3), cmap='gray', vmin=0, vmax=255);
```

2. เขียนฟังก์ชัน `flip(img)` สำหรับพลิกรูป `img`ตามแกนตั้ง

```
def flip(img):
    # img: the numpy array of an image
    # out: img flipped with respect to vertical axis

    out = img.copy()
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            out[i,j] = img[i,img.shape[1] - 1 - j]
    
    return out
```

```
plt.imshow(flip(img), cmap='gray', vmin=0, vmax=255);
```

3. เขียนฟังก์ชัน `rotate(img)` สำหรับการหมุน `img` ทวนเข็มไป 90 องศา

```
def rotate(img):
    # img: the numpy array of an image
    # out: img rotated 90 degree counterclockwisely

    out = np.zeros((img.shape[1], img.shape[0]))
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            out[j,i] = img[i, img.shape[1] - 1 - j]
    
    return out
```

```
plt.imshow(rotate(img), cmap='gray', vmin=0, vmax=255);
```

```
# plot เวกเตอร์โดยใช้ฟังก์ชัน plt.quiver(0,0,*vector)
# ความหมายของ *args คือจะใส่ตัวแปรกี่ตัวก็ได้ 
# โดยที่ตัวแปรที่ใส่ทั้งหมดจะถูกเก็บไว้ใน tuple ที่ชื่อว่า args

def plot_vector(*args):
  plt.xlim(-5,5) # ตั้งค่าต่ำสุดและสูงสุดของ x ที่จะวาดออกมา
  plt.ylim(-5,5) # ตั้งค่าต่ำสุดและสูงสุดของ y ที่จะวาดออกมา
  for v in args:
    plt.quiver([0],[0], [v[0]],[v[1]], units='xy', scale=1) #วาดเวกเตอร์
  plt.grid()                    #วาด grid
  plt.gca().set_aspect('equal') #ตั้งสัดส่วนระหว่างแกน x และ y ให้เท่ากัน
```

```
plot_vector([3,4],[4,3])
```

###Exercise 2

เขียนฟังก์ชันชื่อ plot_rotate(u,r) ที่มี argument สองตัวดังนี้: 
- u คือเวกเตอร์ที่มีสองมิติ 
- r คือขนาดมุมที่ต้องการหมุน (ใน radian)

output: พล็อตเวกเตอร์สองตัว ตัวแรกคือเวกเตอร์ u และตัวที่สองคือเวกเตอร์  w ที่เกิดจากการหมุน u ไป r radian สูตรคือ 

w = [[cos(r) , -sin(r)]u<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;[sin(r) , cos(r)]]

```
from math import sin, cos

def plot_rotate(u,r):
    rotater = np.array([[cos(r), -sin(r)],
                        [sin(r), cos(r)]])
    
    w = rotater @ u

    plt.xlim(-5,5)
    plt.ylim(-5,5)
    plt.quiver([0],[0], [u[0]],[u[1]], units='xy', scale=1)  # u
    plt.quiver([0],[0], [w[0]],[w[1]], units='xy', scale=1)  # u
    plt.grid()
    plt.gca().set_aspect('equal')
```

ลองทดสอบฟังก์ชันด้วยการรันโค้ดด้านล่างนี้

```
pi = np.pi # ค่า pi

plot_rotate([3,4],pi/2)
```
