```
from random import random
from math import sqrt, log, sin, cos, pi
import matplotlib.pyplot as plt
import numpy as np

def generateNormal(R1, R2):
    z1 = sqrt(-2*log(R1))*cos(2*pi*R2)
    z2 = sqrt(-2*log(R1))*sin(2*pi*R2)
    return z1, z2

def plotNormal(randoms, title):
    plt.figure(figsize=(16,9))
    plt.hist(randoms, bins=15)
    plt.title(title)
    plt.show()


randoms = [] # list contain normal random variables
n = 50

for i in range( (n+1) // 2 ):
    R1 = random()
    R2 = random()
    N1, N2 = generateNormal(R1, R2)
    randoms.append(N1)
    randoms.append(N2)

randoms = randoms[:n]
title = f"{n} random numbers with normal distribution"
plotNormal(randoms, title)
```
