```
import random as rnd
import math
import matplotlib.pyplot as plt

def generatePoisson(Lambda):
    Threshold = math.exp(-Lambda)
    count = 0
    R = rnd.random() # Random Number [0, 1)

    while R >= Threshold:
        count += 1
        R *= rnd.random()

    return count

def plotPoisson(randoms, title):
    def counter(randoms):    
        counter = {i: 0 for i in range(1+max(randoms))}

        for R in randoms:
            counter[R] += 1
        
        return counter

    counter = counter(randoms)
    width = 1
    plt.figure(figsize=(16,9))
    plt.bar(counter.keys(), counter.values(), width)
    plt.xticks(range(0, 1+max(randoms)))
    plt.yticks(range(1, 1+max(counter.values())))
    plt.title(title)
    plt.show()

randoms = [] # list contain poisson random variables
Lambda = 8
n = 50
for i in range(n):
    randoms.append(generatePoisson(Lambda))

title = f"{n} random numbers with poisson distribution (lambda={Lambda})"
plotPoisson(randoms, title)
```
